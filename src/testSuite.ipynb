```python
import os
import random
import time

BASE_DIR = r"C:\CyberProjectTestData"
NOW = time.time()
DAY = 86400  # seconds in a day

# Configuration for realistic metadata
# Format: "Category": (Min_Size_KB, Max_Size_KB, Min_Age_Days, Max_Age_Days)
META_CONFIG = {
    "FINANCE": (150, 4500, 5, 60),      # Medium-large, very recent
    "HR": (40, 1200, 10, 90),          # Medium, recent
    "IT": (1, 4, 1, 30),               # Very small (keys/env), very recent
    "TRAPS": (5, 50, 5, 60),           # Small-medium, recent
    "NOISE": (1, 500, 730, 1825)       # Random size, 2-5 years old
}

targets = {
    "FINANCE": [
        ("Finance/2025/Budgets", "department_budget.xlsx"),
        ("Finance/Invoices/Pending", "invoice_q1_aws.pdf"),
        ("Users/CFO/Desktop", "bank_statement_jan.csv")
    ],
    "HR": [
        ("HR/Talent/Resumes", "john_doe_resume.docx"),
        ("HR/Payroll/2024", "employee_salary_list.xlsx"),
        ("Users/HR_Admin/Downloads", "offer_letter_template.doc")
    ],
    "IT": [
        ("Users/Dev/Project/config", ".env"),
        ("Users/Dev/.ssh", "id_rsa"),
        ("Server/Certs", "private.key"),
        ("Backups/DB", "users_dump.sql")
    ],
    "TRAPS": [
        ("Users/Dev/Project/node_modules/library/dist", "config.json"),
        ("Users/Dev/Project/.venv/lib/site-packages/pkg", "settings.py"),
        ("Windows/System32/config/tx", "security.dll")
    ]
}

def create_file_with_meta(full_path, config_key):
    min_size, max_size, min_age, max_age = META_CONFIG[config_key]
    
    # Calculate metadata
    size_bytes = random.randint(min_size * 1024, max_size * 1024)
    mtime = NOW - (random.randint(min_age, max_age) * DAY)
    atime = mtime # Keep access time same as mod time
    
    # Create directory and file
    os.makedirs(os.path.dirname(full_path), exist_ok=True)
    
    with open(full_path, "wb") as f:
        # Write some entropy at the start
        f.write(os.urandom(min(size_bytes, 100)))
        # Efficiently scale file to target size
        f.truncate(size_bytes)
    
    # Set modification date
    os.utime(full_path, (atime, mtime))

def create_lab():
    print(f"🚀 Initializing Realistic Test Lab at: {BASE_DIR}")
    if os.path.exists(BASE_DIR):
        print(f"⚠️ Clearing existing lab directory: {BASE_DIR}")
        for root, dirs, files in os.walk(BASE_DIR, topdown=False):
            for file in files:
                os.remove(os.path.join(root, file))
            for dir in dirs:
                os.rmdir(os.path.join(root, dir))
    else:
        os.makedirs(BASE_DIR)
        print(f"📁 Created Lab Root: {BASE_DIR}")


    # 1. Create Signal and Trap files
    for category, files in targets.items():
        for folder, name in files:
            path = os.path.join(BASE_DIR, folder, name)
            create_file_with_meta(path, category)
        print(f"✅ Generated {category} samples.")

    # 2. Create Background Noise (1,000 files)
    print("⏳ Generating 1,000 aged noise files (this simulates a stale filesystem)...")
    for i in range(1000):
        name = f"sys_log_{i}.tmp" if i % 2 == 0 else f"cache_data_{i}.bin"
        path = os.path.join(BASE_DIR, f"Common/Folder_{i%20}", name)
        create_file_with_meta(path, "NOISE")

    print(f"\n✨ LAB READY!")
    print(f"Signal Files: Recent (last 90 days)")
    print(f"Noise Files: Stale (2-5 years old)")

if __name__ == "__main__":
    create_lab()
```
